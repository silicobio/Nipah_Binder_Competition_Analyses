# Download Structures from Proteinbase

```python
import pandas as pd
from urllib.request import urlretrieve
```

```python
## Load in the previously processed collection data
collection_df = pd.read_csv('../data/proteinbase_collection_nipah-binder-competition-all-submissions_processed.csv')
```

```python
## Loop through each row and download structure file
for index, row in collection_df.iterrows():
    id = row['id']
    structure_url = row['boltz2_structure_prediction']
    print(f'Downloading structure for {id} from {structure_url}...')
    filename = f'../data/structures/{id}_boltz2_complex.cif'
    urlretrieve(structure_url, filename)
```
